```python
import pandas as pd
import math
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import preprocessing

label_encoder = preprocessing.LabelEncoder()

df = pd.read_csv(r"Training_Input_2dx8C9Q.csv")
pd.set_option('display.max_columns', 100)
df.head()
```

```python
df2 = pd.read_csv(r"Training_Output_IJhBXtA.csv")
df_test = pd.read_csv(r"Testing_Input_dPKY3Rf.csv")
train = pd.merge(df, df2, left_index=True, right_index=True)

train.loc[df.weight_I_1 > 0, 'weight_I_1'] = 1
train.loc[df.weight_I_2 > 0, 'weight_I_2'] = 1
train.loc[df.weight_I_3 > 0, 'weight_I_3'] = 1
train.loc[df.weight_I_4 > 0, 'weight_I_4'] = 1
train.loc[df.weight_I_5 > 0, 'weight_I_5'] = 1
train.loc[df.weight_I_6 > 0, 'weight_I_6'] = 1
train.loc[df.weight_I_7 > 0, 'weight_I_7'] = 1

train['weight_I_1'] = train['weight_I_1'].astype(int)
train['weight_I_2'] = train['weight_I_2'].astype(int)
train['weight_I_3'] = train['weight_I_3'].astype(int)
train['weight_I_4'] = train['weight_I_4'].astype(int)
train['weight_I_5'] = train['weight_I_5'].astype(int)
train['weight_I_6'] = train['weight_I_6'].astype(int)
train['weight_I_7'] = train['weight_I_7'].astype(int)

train.head()
```

```python
df_test.head()
```

```python
train['weight_comb'] = train[['weight_I_1','weight_I_2']].dot([10,1])
train['weight_comb'] = train[['weight_comb','weight_I_3']].dot([10,1])
train['weight_comb'] = train[['weight_comb','weight_I_4']].dot([10,1])
train['weight_comb'] = train[['weight_comb','weight_I_5']].dot([10,1])
train['weight_comb'] = train[['weight_comb','weight_I_6']].dot([10,1])
train['weight_comb'] = train[['weight_comb','weight_I_7']].dot([10,1])
                             
train.head()
```

```python
for i in range(1,4):
    train['X_' + str(i)+'_lag_mean'] = train.filter(regex='X_'+str(i)+'_lag_').apply(lambda row : row.mean(), axis = 1)
    
for i in range(1,8):
    train['I_' + str(i)+'_lag_mean'] = train.filter(regex='I_'+str(i)+'_lag_').apply(lambda row : row.mean(), axis = 1)

train.head()
```

```python
df_comb = pd.DataFrame()

df_comb['X_1_lag_mean'] = train['X_1_lag_mean'].astype(float)
df_comb['X_2_lag_mean'] = train['X_2_lag_mean'].astype(float)
df_comb['X_3_lag_mean'] = train['X_3_lag_mean'].astype(float)
df_comb['I_1_lag_mean'] = train['I_1_lag_mean'].astype(float)
df_comb['I_2_lag_mean'] = train['I_2_lag_mean'].astype(float)
df_comb['I_3_lag_mean'] = train['I_3_lag_mean'].astype(float)
df_comb['I_4_lag_mean'] = train['I_4_lag_mean'].astype(float)
df_comb['I_5_lag_mean'] = train['I_5_lag_mean'].astype(float)
df_comb['I_6_lag_mean'] = train['I_6_lag_mean'].astype(float)
df_comb['I_7_lag_mean'] = train['I_7_lag_mean'].astype(float)
df_comb['weight_comb'] = train['weight_comb']
df_comb['Target'] = train['Target'].astype(float)
df_comb.head()
```

```python
plt.figure(figsize=(18,18))
plt.rcParams["axes.labelsize"] = 20
sns.set(font_scale=1.4)
sns.heatmap(df_comb.corr(), annot = True ,linewidths=.1)
plt.show()
```

```python
X_train = df_comb.drop(['Target'],1)
X_train.head()
```

```python
y_train =  df_comb['Target']
y_train.head()
```

```python
y_train_encoded = label_encoder.fit_transform(y_train)
print(y_train_encoded)
```

```python
df_test.loc[df.weight_I_1 > 0, 'weight_I_1'] = 1
df_test.loc[df.weight_I_2 > 0, 'weight_I_2'] = 1
df_test.loc[df.weight_I_3 > 0, 'weight_I_3'] = 1
df_test.loc[df.weight_I_4 > 0, 'weight_I_4'] = 1
df_test.loc[df.weight_I_5 > 0, 'weight_I_5'] = 1
df_test.loc[df.weight_I_6 > 0, 'weight_I_6'] = 1
df_test.loc[df.weight_I_7 > 0, 'weight_I_7'] = 1

df_test['weight_I_1'] = df_test['weight_I_1'].astype(int)
df_test['weight_I_2'] = df_test['weight_I_2'].astype(int)
df_test['weight_I_3'] = df_test['weight_I_3'].astype(int)
df_test['weight_I_4'] = df_test['weight_I_4'].astype(int)
df_test['weight_I_5'] = df_test['weight_I_5'].astype(int)
df_test['weight_I_6'] = df_test['weight_I_6'].astype(int)
df_test['weight_I_7'] = df_test['weight_I_7'].astype(int)

df_test['weight_comb'] = df_test[['weight_I_1','weight_I_2']].dot([10,1])
df_test['weight_comb'] = df_test[['weight_comb','weight_I_3']].dot([10,1])
df_test['weight_comb'] = df_test[['weight_comb','weight_I_4']].dot([10,1])
df_test['weight_comb'] = df_test[['weight_comb','weight_I_5']].dot([10,1])
df_test['weight_comb'] = df_test[['weight_comb','weight_I_6']].dot([10,1])
df_test['weight_comb'] = df_test[['weight_comb','weight_I_7']].dot([10,1])

df_test.head()
```

```python
for i in range(1,4):
    df_test['X_' + str(i)+'_lag_mean'] = df_test.filter(regex='X_'+str(i)+'_lag_').apply(lambda row : row.mean(), axis = 1)
    
for i in range(1,8):
    df_test['I_' + str(i)+'_lag_mean'] = df_test.filter(regex='I_'+str(i)+'_lag_').apply(lambda row : row.mean(), axis = 1)

df_test.head()
```

```python
X_test = pd.DataFrame()

X_test['X_1_lag_mean'] = df_test['X_1_lag_mean'].astype(float)
X_test['X_2_lag_mean'] = df_test['X_2_lag_mean'].astype(float)
X_test['X_3_lag_mean'] = df_test['X_3_lag_mean'].astype(float)
X_test['I_1_lag_mean'] = df_test['I_1_lag_mean'].astype(float)
X_test['I_2_lag_mean'] = df_test['I_2_lag_mean'].astype(float)
X_test['I_3_lag_mean'] = df_test['I_3_lag_mean'].astype(float)
X_test['I_4_lag_mean'] = df_test['I_4_lag_mean'].astype(float)
X_test['I_5_lag_mean'] = df_test['I_5_lag_mean'].astype(float)
X_test['I_6_lag_mean'] = df_test['I_6_lag_mean'].astype(float)
X_test['I_7_lag_mean'] = df_test['I_7_lag_mean'].astype(float)
X_test['weight_comb'] = df_test['weight_comb']

X_test.head()
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression().fit(X_train, y_train_encoded)
print(reg.score(X_train, y_train_encoded))
predictions = reg.predict(X_test)
X_test['prediction'] = predictions
X_test.head()
```
